### Q1) Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['amount_in_usd'], data=df)
plt.title('Boxplot on amount_in_usd')
plt.show()
```

```python
# calculate interquartile range
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)

IQR = Q3 - Q1

# calculate the outlier cutoff
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# identify outliers
outliers = df[(df['amount_in_usd'] < lower_bound) | (df['amount_in_usd'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# replace outliers with median value
median = df['amount_in_usd'].median()   
df['amount_in_usd'] = np.where((df['amount_in_usd'] < lower_bound) | (df['amount_in_usd'] > upper_bound), median, df['amount_in_usd'])
```

### Q4) Try to extract new features by grouping existing columns

```python
# to get influential features

def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt

lags = 7 
ffday = 1
customer_id_col = 'CUSTOMER_NUMBER'
create_date_col = 'ORDER_CREATION_DATE'
net_amount_col = 'amount_in_usd'
df = difference_in_days(df,lags,ffday,customer_id_col, create_date_col, net_amount_col)
df
```

### Q2) Try different machine learning models like - Linear Regression, Support Vector Machine, Decision Tree, Random Forest, AdaBoost, Xgboost etc. 


```python
test_size = 0.2
seed = 42

# load dataset and split into train and test sets
# selected_columns = ['CUSTOMER_ORDER_ID', 'PURCHASE_ORDER_TYPE', 'amount_in_usd','CREDIT_STATUS', 'SOLD_TO_PARTY', 'DIVISION']
X = df[['RELEASED_CREDIT_VALUE','PURCHASE_ORDER_TYPE','CREDIT_STATUS','unique_cust_id','Last-1day_Sales','Last-2day_Sales','Last-3day_Sales','Last-1day_Diff','Last-2day_Diff','Last-3day_Diff']]
y = df['ORDER_AMOUNT']

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
# store model results
results = pd.DataFrame(columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Score'])
```

```python
X_train
```

```python
# function to train model and evaluate model
def regression_model(model_name, model):
    # load model
    regressor = model
    
    # training model
    regressor.fit(X_train, y_train)
    
    # predicting model
    y_pred = regressor.predict(X_test)
    
    # calculating model performance
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    r2score = r2_score(y_test, y_pred)
    
    # store model results
    results.loc[len(results)] = [model_name, mae, mse, rmse, r2score]
        
    # print model results
    print('Model:', model_name)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('R2 Score:', metrics.r2_score(y_test, y_pred))
    
```

### Q6) Perform Hyperparameter tuning, select best hyperparameters by using appropriate algorithms


```python
# function to use random search to find best hyperparameters
def RandomSearch(model, X_train, y_train, X_test, y_test):
    
    # define parameters
    random_param = {
        'learning_rate': [0.1, 0.01],
        'n_estimators': [20, 40, 60, 80, 100]
    }
    
    # load model
    model_ = model

    # random search applying 5 fold cross validation
    random_search = RandomizedSearchCV(model_, param_distributions=random_param, n_iter=10, scoring='r2', cv=5,verbose=1, random_state=42)
    random_search.fit(X_train, y_train)
    
    # print best parameters and best score
    print(random_search.best_params_)
    print(random_search.best_score_)
    print(random_search.best_estimator_)
    
    # model evaluation
    y_pred = random_search.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # print model evaluation
    print('MSE: ', mse)
    print('RMSE: ', rmse)
    print('R2: ', r2)
    
    return random_search.best_estimator_, mse, rmse, r2
    
    
```

```python
best_estimator_r, mse_r, rmse_r, r2_r = RandomSearch(XGBRegressor(), X_train, y_train, X_test, y_test)
```

```python
# function to use grid search to find best hyperparameters
def GridSearch(model, X_train, y_train, X_test, y_test):
    # define parameters
    random_grid = {
        'learning_rate': [0.1, 0.01],
        'n_estimators': [20, 40, 60, 80, 100]
    }
    
    # load model
    model_ = model
    
    # grid search applying 5 fold cross validation
    grid_search = GridSearchCV(estimator = model_, param_grid = random_grid, cv = 5, n_jobs = -1, verbose = 5)
    
    # fit model
    grid_search.fit(X_train, y_train)
    
    # print best parameters and best score
    print(grid_search.best_params_)
    print(grid_search.best_score_)
    print(grid_search.best_estimator_)
    
    # model evaluation
    y_pred = grid_search.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # print model evaluation
    print('MSE: ', mse)
    print('RMSE: ', rmse)
    print('R2: ', r2)
    
    return grid_search.best_estimator_, mse, rmse, r2
```

```python
best_estimator_g, mse_g, rmse_g, r2_g = RandomSearch(XGBRegressor(), X_train, y_train, X_test, y_test)
```
